```python
import numpy as np
import os
from tensorflow.keras.utils import load_img, img_to_array
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
img_dir = 'flag_data'
batch_size = 16
img_height = 224
img_width = 224
```

```python
train_ds = tf.keras.utils.image_dataset_from_directory(
  img_dir,
  validation_split=0.2,
  subset="training",
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size)
```

```python
val_ds = tf.keras.utils.image_dataset_from_directory(
  img_dir,
  validation_split=0.2,
  subset="validation",
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size)
```

```python
class_names = train_ds.class_names
```

```python
plt.figure(figsize=(10, 10))
for images, labels in train_ds.take(1):
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(class_names[labels[i]])
        plt.axis("off")
```

```python
num_classes = len(class_names)

model = tf.keras.Sequential([
  tf.keras.layers.Rescaling(1./255),
  tf.keras.layers.Conv2D(32, 3, activation='relu'),
  tf.keras.layers.MaxPooling2D(),
  tf.keras.layers.Conv2D(32, 3, activation='relu'),
  tf.keras.layers.MaxPooling2D(),
  tf.keras.layers.Conv2D(32, 3, activation='relu'),
  tf.keras.layers.MaxPooling2D(),
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dense(num_classes, activation=tf.keras.activations.softmax)
])
```

```python
model.compile(
  optimizer='adam',
  loss=tf.keras.losses.SparseCategoricalCrossentropy(),
  metrics=['accuracy'])
```

```python
model.fit(
  train_ds,
  validation_data=val_ds,
  epochs=3
)
```

```python
def predict_image(filename):
    test_image = load_img(filename, target_size=(img_height, img_width))
    plt.imshow(test_image)
    test_image = img_to_array(test_image)
    test_image = test_image.reshape(1, img_height, img_width, 3)
    test_image = test_image.astype('float32')
    pred = model.predict(test_image)
    probs = np.sort(pred[0])[-10:][::-1]
    k = len(probs[probs>=0.1])
    top_index, top_prob = np.argsort(pred[0])[-k:][::-1], np.sort(pred[0])[-k:][::-1]
    for i in range(k):
        img = load_img(f'flags/{class_names[top_index[i]]}.gif', target_size=(img_height, img_width))
        plt.imshow(img)
        break
    print(top_prob)
    print([class_names[i] for i in top_index])
```
